```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

#  MSE MAE  ( )
mse_list = [1.435, 1.925, 2.169, 1.443, 2.139, 2.847, 3.006, 2.077, 2.497, 7.382, 5.137, 4.724]
mae_list = [0.745, 0.903, 1.041, 0.797, 0.931, 1.144, 1.161, 0.914, 1.004, 2.003, 1.544, 1.445]

#  
models = ['MIFlu', 'GPT2', 'DLinear', 'PatchTST', 'TimesNet', 'FEDFormer', 'Autoformer', 'Sationary', 'ETSformer', 'LightTS', 'Informer', 'Reformer']

# Data preparation
data = {
    'Model': np.repeat(models, 2),
    'Metric': ['MSE', 'MAE'] * len(models),
    'Value': [val for pair in zip(mse_list, mae_list) for val in pair]
}

# Convert to DataFrame
df = pd.DataFrame(data)

# Create the bar plot
plt.figure(figsize=(14, 8))  # Adjust figure size
sns.barplot(x='Model', y='Value', hue='Metric', data=df, palette='Set2')

# Increase the font size
plt.rcParams.update({'font.size': 14})

# Add title and labels
plt.ylabel('Value', fontsize=14)
plt.xlabel('Model', fontsize=14)

# Adjust tick label font size
plt.tick_params(axis='both', which='major', labelsize=12)

# Add legend
plt.legend(title='Metric', bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=12)

# Show the plot
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
# plt.savefig('bargraph.png', dpi=700, bbox_inches='tight')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

metric_list = [1.435, 1.925, 2.169, 1.443, 2.139, 2.847, 3.006, 2.077, 2.497, 7.382, 5.137, 4.724,
             0.745, 0.903, 1.041, 0.797, 0.931, 1.144, 1.161, 0.914, 1.004, 2.003, 1.544, 1.445
            ]

# Synthetic data for demonstration purposes
lead_times = ['MSE', 'MAE']
models = ['MIFlu', 'GPT4TS', 'DLinear', 'PatchTST', 'TimesNet', 'FEDFormer', 'Autoformer', 'Sationary', 'ETSformer', 'LightTS', 'Informer', 'Reformer']
data = {
    'Lead Time (weeks)': np.repeat(lead_times, len(models)),
    'Model': models * len(lead_times),
    'Metric': metric_list  # Use CORR instead of RMSE
}

# Convert to DataFrame
import pandas as pd
df = pd.DataFrame(data)

# Create the bar plot
plt.figure(figsize=(10, 6))  # Smaller figure size
sns.barplot(x='Lead Time (weeks)', y='Metric', hue='Model', data=df, palette='Set2')

# Increase the font size
plt.rcParams.update({'font.size': 12})

# Add title and labels
plt.ylabel('Value', fontsize=14)
plt.xlabel('Evaluation Metric', fontsize=14)

# Adjust tick label font size
plt.tick_params(axis='both', which='major', labelsize=12)

# Adjust y-axis to better visualize differences
# plt.ylim(0.82, 0.97)

# Display the plot
plt.legend(title='Model', bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=12)
# plt.title('Comparison of Average Value of CORR Across Methods and Horizons', fontsize=16)

plt.savefig('bargraph.png', dpi=700, bbox_inches='tight')
# plt.show()
```
