```python
import pandas as pd

data = pd.read_csv('../data/ILI/ILI_region_feature_2020-2024.csv')
```

```python
data
```

```python

# Pivot the data to have each Region as a column with ILITOTAL values for each week without cumulative sum
weekly_data = data.pivot_table(values="ILITOTAL", index=["YEAR", "WEEK"], columns="REGION", aggfunc="sum")

# Fill NaN values with 0 (assuming missing values mean no cases reported)
weekly_data = weekly_data.fillna(0)

# Reset index to flatten the DataFrame
weekly_data = weekly_data.reset_index()
```

```python
weekly_data.to_csv('ILI_region_feature_2020-2024_.csv')
```
